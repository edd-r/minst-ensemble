# Visualising Minst data

```python
import os
import sys
import importlib
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# add project modules to the path
path_to_module = os.path.abspath(os.path.join(os.getcwd(), "..", "src/"))
sys.path.append(path_to_module)

import src.models.train_model as train_model
import src.visualization.visualize as vis

plt.rcParams['figure.figsize'] = [10, 5]

%matplotlib notebook
```

load in the data from each different method of processing it

```python
path_to_data = os.path.abspath(os.path.join(os.getcwd(),
                                            "..",
                                            "data/processed/"
                                           ))

scaled_data_dict = train_model.load_processed_data(file_path=path_to_data,
                                            augmented_suffix="",
                                            blurred_suffix="",
                                            )

aug_scaled_data_dict = train_model.load_processed_data(file_path=path_to_data,
                                                blurred_suffix="",
                                            )

blur_scaled_data_dict = train_model.load_processed_data(file_path=path_to_data,
                                                augmented_suffix="",
                                            )

aug_blur_scaled_data_dict = train_model.load_processed_data(file_path=path_to_data)




images_dict_list =(scaled_data_dict,
              aug_scaled_data_dict,
              blur_scaled_data_dict,
              aug_blur_scaled_data_dict)

```

process the training data (we don't want to cheat!) to compute the mean image 

```python
images_list = [vis.average_image_by_class(data["X_train"], data["y_train"])
               for data in images_dict_list
              ]

image_categories =("Scaled",
                   "Augmented, scaled",
                   "Blurred, scaled",
                   "Augmented, blurred, scaled",
                  )
```

now look at the images...

```python
fig, axes = vis.compare_images(images_list, image_categories)
```

ok, so we can see some of the difference between the different datasets here. 
- blurred is slighly fainter than the scaled images, but not too much difference
- augmeneted; we see the jitter come through clearly
- augmented and blurred again makes it harder to distiguish the numbers.

Nothing surprising, but it is interesting to see the differences
